## Question 2. Produce a histogram of the minimum ages of the ant specimens in this dataset (15 pts) . The name of the X-Axis should be "Age of Fossil (millions of years)". 

```python
import matplotlib.pyplot as plt
import plotnine as p9
import dendropy
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import plotnine as p9
myplot = (p9.ggplot(data=fossil, 
                    mapping=p9.aes(x='min_ma')) +
              p9.geom_histogram(binwidth=1))
my_plt = myplot.draw()
p9_ax = my_plt.axes[0]
p9_ax.set_xlabel("Age of Fossil (millions of years)", size=14)
plt.show()
```

## Question 3. Produce a bar plot of how many fossils exist for each subfamily. (10 pts). 

```python
mybarplot = (p9.ggplot(data=fossil, 
                    mapping=p9.aes(x='subfamily')) +
                    p9.geom_bar()+
                    p9.theme(axis_text_x = p9.element_text(angle=90))
)
my_bar_plot = mybarplot.draw()
p9_ax = my_bar_plot.axes[0]
p9_ax.set_xlabel("Subfamily", size=14)
plt.show()
```

# Part Two: Genetic Data

In the data directory, there is an ~unknown~ DNA sequence in a FASTA file. 

## Question 4: Perform a BLAST search of this ~unknown~ sequence (5 pts)

```python
from Bio.Blast import NCBIWWW
from Bio import SeqIO
randomfile = SeqIO.read("data/mysterSequence.fasta", format="fasta")
random = NCBIWWW.qblast("blastn", "nt", randomfile.format("fasta"))
```

```python
sequence = random
#print(sequence.read())
```

## Question 5: Create a results folder and write your sequence to it. Not all of this question needs to be completed in python (5 pts)

```python
save_file = open("maybeempty3", "w")
save_file.write(sequence.read())
save_file.close()
```

## Question 6: What type of organism do you think these data came from? (5pts)

```python
from Bio.Blast import NCBIXML
parsed = NCBIXML.read(open("results/maybeempty3"))
```

```python
alignment_set = parsed.alignments
for individual_alignment in alignment_set:
    print(individual_alignment)
```

These data came from ants from searching the scientific names from the printed individual alignment

## Question 7: What is one analysis you might perform to confirm that these BLAST hits are real? (5pts)

A significant E-Value will confirm the BLAST results with a number that is closest to 0. 

## Question 8: Perform the analysis you suggest in Q7. (10pts) 

```python
for individual_alignment in alignment_set:
    print(individual_alignment.hsps[0].expect)
#values must be less than 0.0001 for significance
```
